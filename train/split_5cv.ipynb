### This script is to split the data into 5-fold cross validation

```python
from os import path
import numpy as np
import glob
import os
import sys
import random


exclude_set = set()
exclude = open("../../data/exclude.txt", 'r')
for line in exclude:
    line = line.strip()
    print(line.split('\t')[1])
    if float(line.split('\t')[1]) > 0.5:
        exclude_set.add(line.split('\t')[0])
        print(line.split('\t')[0])

def add_train(S, train, exclude_set):
    '''
    input: set to be added; file to be written; set in which the sample need to be excludedd
    output: write set into the file, excluding samples in exclide set
    '''
    for i in S:
        if i in exclude_set:
            print('excluded')
        else:
            if path.exists('../../data/label/'+i):
                train.write(('%s\t%s\n') % ('../../data/raw_image/mace_full_png/'+i, '../../data/label/'+i))
            else:
                train.write(('%s\t%s\n') % ('../../data/raw_image/mace_full_png/'+i, 'neg' ))

                
#split the whole dataset into 5 subset with same size
set_one = set()
set_two = set()
set_three = set()
set_four = set()
set_five = set()

mace_full = list()
for filename in os.listdir('../../data/raw_image/mace_full_png'):
    mace_full.append(filename)
random.shuffle(mace_full)

count = 0
for i in mace_full:
    count +=1
    if count <= 7545*1:
        set_one.add(i)
    elif count <= 7545*2:
        set_two.add(i)
    elif count <= 7545*3:
        set_three.add(i)
    elif count <= 7545*4:
        set_four.add(i)
    else:
        set_five.add(i)
```

```python
train_1  = open('train_1_cv.dat', 'w')
test_1  = open('test_1_cv.dat', 'w')

for i in set_one:# add to test set for the cv 1
    test_1.write('../../data/raw_image/mace_full_png/'+i+'\n')
# add to train set for the cv 1 
add_train(set_two, train_1, exclude_set)
add_train(set_three, train_1, exclude_set)
add_train(set_four, train_1, exclude_set)
add_train(set_five, train_1, exclude_set)
```

```python
train_2  = open('train_2_cv.dat', 'w')
test_2  = open('test_2_cv.dat', 'w')

for i in set_two:
    test_2.write('../../data/raw_image/mace_full_png/'+i+'\n')
add_train(set_one, train_2, exclude_set)
add_train(set_three, train_2, exclude_set)
add_train(set_four, train_2, exclude_set)
add_train(set_five, train_2, exclude_set)
```

```python
train_3  = open('train_3_cv.dat', 'w')
test_3  = open('test_3_cv.dat', 'w')

for i in set_three:
    test_3.write('../../data/raw_image/mace_full_png/'+i+'\n')
add_train(set_one, train_3, exclude_set)
add_train(set_two, train_3, exclude_set)
add_train(set_four, train_3, exclude_set)
add_train(set_five, train_3, exclude_set)
```

```python
train_4  = open('train_4_cv.dat', 'w')
test_4  = open('test_4_cv.dat', 'w')

for i in set_four:
    test_4.write('../../data/raw_image/mace_full_png/'+i+'\n')
add_train(set_one, train_4, exclude_set)
add_train(set_two, train_4, exclude_set)
add_train(set_three, train_4, exclude_set)
add_train(set_five, train_4, exclude_set)
```

```python
train_5  = open('train_5_cv.dat', 'w')
test_5  = open('test_5_cv.dat', 'w')
for i in set_five:
    test_5.write('../../data/raw_image/mace_full_png/'+i+'\n')
add_train(set_one, train_5, exclude_set)
add_train(set_two, train_5, exclude_set)
add_train(set_three, train_5, exclude_set)
add_train(set_four, train_5, exclude_set)
```
