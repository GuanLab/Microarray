### This script is made for Dice coefficient calculation.  Read eva_cv1.txt, ..., eva_cv5.txt and calculate the average dice for positive cases for predidction in each fold of cross validation

#### author: Yanan Qin

```python
import nibabel as nib
import numpy as np
import glob
import os
import sys
import component_cut
import cv2
from os import path
import matplotlib.pyplot as plt

import unet
import tensorflow as tf
from keras import backend as K

from keras.backend.tensorflow_backend import set_session
```

```python
counter = 0
avg_accuracy = 0
FILE=open('eva_cv1.txt')  #for the first fold 
for line in FILE:
    line=line.rstrip()
    img_path = line.split('\t')[0]
    accuracy = float(line.split('\t')[1])
#     identity = line.split('\t')[-1]
    img_name = img_path.split('/')[-1]
    label_path = '../../data/label/' + img_name  # only consider the dice index for positive cases
    if path.exists(label_path):
        #positive case
        counter += 1
        avg_accuracy += accuracy
print('cv1', avg_accuracy /counter)        
        
    
counter = 0
avg_accuracy = 0
FILE=open('eva_cv2.txt') #for the second fold
for line in FILE:
    line=line.rstrip()
    img_path = line.split('\t')[0]
    accuracy = float(line.split('\t')[1])
    img_name = img_path.split('/')[-1]
    label_path = '../../data/label/' + img_name
    if path.exists(label_path):
        #positive case
        counter += 1
        avg_accuracy += accuracy
print('cv2', avg_accuracy /counter)        
        
    
counter = 0
avg_accuracy = 0             
FILE=open('eva_cv3.txt') #for the third fold
for line in FILE:
    line=line.rstrip()
    img_path = line.split('\t')[0]
    accuracy = float(line.split('\t')[1])
    img_name = img_path.split('/')[-1]
    label_path = '../../data/label/' + img_name
    if path.exists(label_path):
        #positive case
        counter += 1
        avg_accuracy += accuracy
print('cv3', avg_accuracy /counter)        
        
    
counter = 0
avg_accuracy = 0       
FILE=open('eva_cv4.txt') #for the firth fold
for line in FILE:
    line=line.rstrip()
    img_path = line.split('\t')[0]
    accuracy = float(line.split('\t')[1])
    img_name = img_path.split('/')[-1]
    label_path = '../../data/label/' + img_name
    if path.exists(label_path):
        #positive case
        counter += 1
        avg_accuracy += accuracy
print('cv4', avg_accuracy /counter)        
        
    
counter = 0
avg_accuracy = 0      
FILE=open('eva_cv5.txt') #for the fifth fold
for line in FILE:
    line=line.rstrip()
    img_path = line.split('\t')[0]
    accuracy = float(line.split('\t')[1])
    img_name = img_path.split('/')[-1]
    label_path = '../../data/label/' + img_name
    if path.exists(label_path):
        #positive case
        counter += 1
        avg_accuracy += accuracy
print('cv5', avg_accuracy /counter)        
    
```
